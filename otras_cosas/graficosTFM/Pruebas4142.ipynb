```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df41a = pd.read_csv('prueba41a.csv',delimiter=',',index_col=3)
df41b = pd.read_csv('prueba41b.csv',delimiter=',',index_col=3)
df41c = pd.read_csv('prueba41c.csv',delimiter=',',index_col=3) #Este tiene fidelidad sin supresión alguna DepolarGauss
df41d = pd.read_csv('prueba41d.csv',delimiter=',',index_col=3) #Este tiene fidelidad sin supresión alguna Dephase
df42a = pd.read_csv('prueba42a.csv',delimiter=',',index_col=3)
df42b = pd.read_csv('prueba42b.csv',delimiter=',',index_col=3)
```

# Con error en canal de Dephase (Z)

```python
#Fidelity
dfGraf = pd.concat([df41a['Mean fidelity'],df42a['Mean fidelity'],
                                         df41d['Mean fidelity']],axis=1)
dfGraf.columns=['Purificación','QEC','Sin corrección']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['Purificación'].plot(color='b',legend='True')
_ = dfGraf['QEC'].plot(color='r',legend='True')
_ = dfGraf['Sin corrección'].plot(color='k',linestyle='dashed',legend='True')

plt.title('Evolución de la fidelidad con distancia \n aplicando técnicas de supresión')
plt.xlabel('Distancia enlace (km)')
plt.ylabel('Fidelidad')
```

```python
#Teleported states
dfGraf = pd.concat([df41a['Teleported states'],df42a['Teleported states'],
                                         df41d['Teleported states']],axis=1)
dfGraf.columns=['Purificación','QEC(Shor)','Sin corrección']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['Purificación'].plot(color='b',legend='True')
_ = dfGraf['QEC(Shor)'].plot(color='r',legend='True')

ax = plt.gca()
ax.set_xlim([0, 100])
ax.set_ylim([0, 300])
plt.title('Número de estados teleportados en 1 segundo')
plt.xlabel('Distancia enlace (km)')
plt.ylabel('#qubits')
```

```python
#Mean time
dfGraf = pd.concat([df41a['Mean time'],df42a['Mean time'],
                                         df41d['Mean time']],axis=1)
dfGraf.columns=['Purificación','QEC(Shor)','Sin corrección']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['Purificación'].plot(color='b',legend='True')
_ = dfGraf['QEC(Shor)'].plot(color='r',legend='True')

plt.title('Tiempo medio de teleportación de un qubit')
plt.xlabel('Distancia enlace (km)')
plt.ylabel('nanosegundos')
```

```python
# Fidelity
dfFidelity = pd.concat([df41a['Mean fidelity'], df42a['Mean fidelity'], df41d['Mean fidelity']], axis=1)
dfFidelity.columns = ['Purification', 'QEC', 'No correction']

# Capacity
dfCapacity = pd.concat([df41a['Teleported states'], df42a['Teleported states'], df41d['Teleported states']], axis=1)
dfCapacity.columns = ['Purification', 'QEC (Shor)', 'No correction']

# Mean time
dfTime = pd.concat([df41a['Mean time'], df42a['Mean time'], df41d['Mean time']], axis=1)
dfTime.columns = ['Purification', 'QEC (Shor)', 'No correction']

# Creating subplots
fig, axs = plt.subplots(1, 3,figsize=(16,5))
plt.rcParams.update({'font.size': 12})
# Plot for Fidelity
dfFidelity['Purification'].plot(ax=axs[0], color='b', legend=True)
dfFidelity['QEC'].plot(ax=axs[0], color='r', legend=True)
dfFidelity['No correction'].plot(ax=axs[0], color='k', linestyle='dashed', legend=True)
axs[0].set_title('Fidelity evolution with distance')
axs[0].set_xlabel('Link distance (km)')
axs[0].set_ylabel('State fidelity')
axs[0].text(0.05,0.95,'a'+')')

# Plot for Capacity
dfCapacity['Purification'].plot(ax=axs[1], color='b', legend=True)
dfCapacity['QEC (Shor)'].plot(ax=axs[1], color='r', legend=True)
axs[1].set_xlim([0, 100])
axs[1].set_ylim([0, 300])
axs[1].set_title('Teleported states per second')
axs[1].set_xlabel('Link distance (km)')
axs[1].set_ylabel('# qubits')
axs[1].text(5,250,'b'+')')

# Plot for Mean time
dfTime['Purification'].plot(ax=axs[2], color='b', legend=True)
dfTime['QEC (Shor)'].plot(ax=axs[2], color='r', legend=True)
axs[2].set_title('Average qubit teleportation time')
axs[2].set_xlabel('Link distance (km)')
axs[2].set_ylabel('Round time (ns)')
axs[2].text(1,3.7e7,'c'+')')

# Adjust layout
#plt.tight_layout()
plt.savefig("teleportcomparison.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
