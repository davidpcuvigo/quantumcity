```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df4350 = pd.read_csv('prueba43_50.csv',delimiter=',')
df4360 = pd.read_csv('prueba43_60.csv',delimiter=',')
df4370 = pd.read_csv('prueba43_70.csv',delimiter=',')
df4380 = pd.read_csv('prueba43_80.csv',delimiter=',')
df4390 = pd.read_csv('prueba43_90.csv',delimiter=',')
df4350.reset_index(inplace=True)
df4350['Valor'] = df4350['valor'] + 1
df4350.set_index('Valor',inplace=True)
df4360.reset_index(inplace=True)
df4360['Valor'] = df4360['valor'] + 1
df4360.set_index('Valor',inplace=True)
df4370.reset_index(inplace=True)
df4370['Valor'] = df4370['valor'] + 1
df4370.set_index('Valor',inplace=True)
df4380.reset_index(inplace=True)
df4380['Valor'] = df4380['valor'] + 1
df4380.set_index('Valor',inplace=True)
df4390.reset_index(inplace=True)
df4390['Valor'] = df4390['valor'] + 1
df4390.set_index('Valor',inplace=True)

df4450 = pd.read_csv('prueba44_50.csv',delimiter=',')
df4460 = pd.read_csv('prueba44_60.csv',delimiter=',')
df4470 = pd.read_csv('prueba44_70.csv',delimiter=',')
df4480 = pd.read_csv('prueba44_80.csv',delimiter=',')
df4490 = pd.read_csv('prueba44_90.csv',delimiter=',')
df4450.reset_index(inplace=True)
df4450['Valor'] = df4450['valor']*4 + 2
df4450.set_index('Valor',inplace=True)
df4460.reset_index(inplace=True)
df4460['Valor'] = df4460['valor']*4 + 2
df4460.set_index('Valor',inplace=True)
df4470.reset_index(inplace=True)
df4470['Valor'] = df4470['valor']*4 + 2
df4470.set_index('Valor',inplace=True)
df4480.reset_index(inplace=True)
df4480['Valor'] = df4480['valor']*4 + 2
df4480.set_index('Valor',inplace=True)
df4490.reset_index(inplace=True)
df4490['Valor'] = df4490['valor']*4 + 2
df4490.set_index('Valor',inplace=True)
```

```python
#Fidelity
dfGraf = pd.concat([df4350['Mean fidel'],df4360['Mean fidel'],
                    df4370['Mean fidel'],df4380['Mean fidel'],df4390['Mean fidel']
                   ],
                   axis=1)

#dfGraf['Valor'] = dfGraf['index']*3.878 + 10

dfGraf.columns=['0.5','0.6','0.7','0.8','0.9']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['0.5'].plot(color='b',legend='True')
_ = dfGraf['0.6'].plot(color='r',legend='True')
_ = dfGraf['0.7'].plot(color='g',legend='True')
_ = dfGraf['0.8'].plot(color='y',legend='True')
_ = dfGraf['0.9'].plot(color='m',legend='True')
plt.title('Evolución de la fidelidad con la distancia extremo-extremo \n 1 switch')
plt.xlabel('km')
plt.ylabel('Fidelidad')
```
