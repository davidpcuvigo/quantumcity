```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df1 = pd.read_csv('prueba49.csv',delimiter=',',index_col=3)
df2 = pd.read_csv('prueba9.csv',delimiter=',',index_col=3)
df3 = pd.read_csv('prueba10.csv',delimiter=',',index_col=3)
```

# Dependencia de la probabilidad de acierto respecto la fidelidad de la fuente teniendo en cuenta ruido

```python
# Plot the results
plt.figure(figsize=(6, 3))
plt.plot(df1['Win p'], color='r',label='1 Request')
plt.axvline(x=1/(2**(1/2)), color='black', linestyle='--', label='Theoretical threshold')
plt.xlabel('Fidelidad de la fuente')
plt.ylabel('Porcentaje de victorias (%)')
plt.title('Probabilidad de acierto en el protocolo CHSH (NV)')
plt.xticks([0,0.2,0.4,0.6,0.8,1] + [1/(2**(1/2))]) #Adding the extra point
current_ticks = list(plt.xticks()[0])  # Get current x-ticks
tick_labels = [f"{tick:.1f}" if tick != 1/(2**(1/2)) else "$\\frac{1}{\\sqrt{2}}$" for tick in current_ticks]

# Set the x-ticks and their labels
plt.xticks(current_ticks, tick_labels)

plt.legend()
#plt.grid(True)
plt.show()
```

```python
# Plot the results
plt.figure(figsize=(6, 3))
plt.plot(df1['Win p'], color='r',label='Simulation results')
plt.axvline(x=1/(2**(1/4)), color='black', linestyle='--', label='Theoretical threshold (75%)')
plt.xlabel('Emission fidelity')
plt.ylabel('Win percentage (%)')
plt.title('Win percentage in CHSH game (NV)')
plt.xticks([0,0.25,0.5,0.75,1] + [1/(2**(1/4))]) #Adding the extra point
current_ticks = list(plt.xticks()[0])  # Get current x-ticks
tick_labels = [f"{tick:.2f}" if tick != 1/(2**(1/4)) else "$\\frac{1}{2^{1/4}}$" for tick in current_ticks]

# Set the x-ticks and their labels
plt.xticks(current_ticks, tick_labels)

plt.legend()
plt.savefig("chshposter.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
