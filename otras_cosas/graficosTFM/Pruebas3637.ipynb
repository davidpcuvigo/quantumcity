```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df36a = pd.read_csv('prueba36a.csv',delimiter=',',index_col=3)
df36b = pd.read_csv('prueba36b.csv',delimiter=',',index_col=3)
df37a = pd.read_csv('prueba37a.csv',delimiter=',',index_col=3)
df37b = pd.read_csv('prueba37b.csv',delimiter=',',index_col=3)
```

```python
#Fidelity
dfGraf = pd.concat([df36a['Mean fidel'],df36b['Mean fidel'],
                                         df37a['Mean fidel'],df37b['Mean fidel']],axis=1)

dfGraf.columns=['Demanda alta - Oldest','Demanda baja - Oldest','Demanda alta - Newest','Demanda baja - Newest']
fig = plt.figure(figsize=(9,4.5))
_ = dfGraf['Demanda alta - Oldest'].plot(color='b',legend='True')
_ = dfGraf['Demanda baja - Oldest'].plot(color='r',legend='True')
_ = dfGraf['Demanda alta - Newest'].plot(color='g',legend='True')
_ = dfGraf['Demanda baja - Newest'].plot(color='y',legend='True')
plt.title('Evolución de la fidelidad con el \n tamaño del búfer de envío')
plt.xlabel('Posiciones de memoria')
plt.ylabel('Fidelidad')
```

```python
#Mean time
dfGraf = pd.concat([df36a['Mean time'],df36b['Mean time'],
                                         df37a['Mean time'],df37b['Mean time']],axis=1)

dfGraf.columns=['Demanda alta - Oldest','Demanda baja - Oldest','Demanda alta - Newest','Demanda baja - Newest']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['Demanda alta - Oldest'].plot(color='b',legend='True')
_ = dfGraf['Demanda baja - Oldest'].plot(color='r',legend='True')
_ = dfGraf['Demanda alta - Newest'].plot(color='g',legend='True')
_ = dfGraf['Demanda baja - Newest'].plot(color='y',legend='True')
plt.axis([10, 100, 530000, 536000])
plt.title('Evolución del tiempo medio de teleportación \n con el tamaño del búfer de envío')
plt.xlabel('Posiciones de memoria')
plt.ylabel('nanosegundos')
```

```python
#Teleported states
dfGraf = pd.concat([df36a['Teleported states'],df36b['Teleported states'],
                                         df37a['Teleported states'],df37b['Teleported states']],axis=1)

dfGraf.columns=['Demanda alta - Oldest','Demanda baja - Oldest','Demanda alta - Newest','Demanda baja - Newest']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['Demanda alta - Oldest'].plot(color='b',legend='True')
_ = dfGraf['Demanda baja - Oldest'].plot(color='r',legend='True')
_ = dfGraf['Demanda alta - Newest'].plot(color='g',legend='True')
_ = dfGraf['Demanda baja - Newest'].plot(color='y',legend='True')
plt.title('Evolución de estados teleportados con \n el tamaño del búfer de envío')
plt.xlabel('Número de posiciones de memoria')
plt.ylabel('Estados teleportados')
```
