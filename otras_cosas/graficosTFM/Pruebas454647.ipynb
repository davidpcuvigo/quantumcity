```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df45 = pd.read_csv('prueba45.csv',delimiter=',',index_col=3)
df46 = pd.read_csv('prueba46.csv',delimiter=',',index_col=3)
df47 = pd.read_csv('prueba47.csv',delimiter=',',index_col=3)
```

```python
#Fidelity
dfGraf = pd.concat([df45['Mean fidelity'],df46['Mean fidelity'],
                                         df47['Mean fidelity']
                                        ],axis=1)
dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolución de la fidelidad con la probabilidad \n de que la fuente emita un estado de Bell')
plt.xlabel('Probabilidad estado Bell')
plt.ylabel('Fidelidad')
```

```python
#Fidelity
dfGraf = pd.concat([df45['Mean time'],df46['Mean time'],
                                         df47['Mean time']
                                        ],axis=1)
dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolución de tiempo medio teleportación con la calidad de a fuente')
plt.xlabel('Probabilidad estado Bell')
plt.ylabel('Nanosegundos')
```

```python
#Fidelity
dfGraf = pd.concat([df45['Teleported states'],df46['Teleported states'],
                                         df47['Teleported states']
                                        ],axis=1)
dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolución estados teleportados con la calidad de la fuente')
plt.xlabel('Probabilidad estado Bell')
plt.ylabel('Nanosegundos')
```
