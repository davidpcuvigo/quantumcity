```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df1 = pd.read_csv('prueba40a.csv',delimiter=',',index_col=3)
df2 = pd.read_csv('prueba40b.csv',delimiter=',',index_col=3)
df3 = pd.read_csv('prueba40c.csv',delimiter=',',index_col=3)
```

# Fidelidad media respecto al número de solicitudes para cada tecnología

```python
average_fidelity1 = df1.groupby('valor')['Mean fidel'].mean().reset_index() #innecesario
average_fidelity2 = df2.groupby('valor')['Mean fidel'].mean().reset_index()
average_fidelity3 = df3.groupby('valor')['Mean fidel'].mean().reset_index()
average_fidelity1.set_index('valor',inplace=True)
average_fidelity2.set_index('valor',inplace=True)
average_fidelity3.set_index('valor',inplace=True)
# Plot the results
plt.figure(figsize=(6, 3))
plt.plot(average_fidelity1['Mean fidel'], color='r',label='1 demanda')
plt.plot(average_fidelity2['Mean fidel'], color='g', label='3 demandas')
plt.plot(average_fidelity3['Mean fidel'], color='b', label='5 demandas')
plt.axvline(x=1900, color='black', linestyle='--', label='Tasa máxima sin pérdidas')
plt.xlabel('Tasa de demanda (Hz)')
plt.ylabel('Fidelidad')
plt.title('Evolución de fidelidad en función \n de la tasa de demanda')
plt.legend()
#plt.grid(True)
plt.show()
```
