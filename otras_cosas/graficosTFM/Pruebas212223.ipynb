```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df21 = pd.read_csv('prueba21.csv',delimiter=',',index_col=3)
df22 = pd.read_csv('prueba22.csv',delimiter=',',index_col=3)
df23 = pd.read_csv('prueba23.csv',delimiter=',',index_col=3)

```

```python
#Fidelity
dfGraf = pd.concat([df21['Mean fidel'],df22['Mean fidel'],
                                         df23['Mean fidel']],axis=1)
#dfGraf.reset_index(inplace=True)
#dfGraf['Valor'] = dfGraf['index']*3.878 + 10
#dfGraf.set_index('Valor',inplace=True)
#dfGraf.columns=['rowold','1 switch','3 switches','5 switches']
dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolución de la fidelidad con el tiempo de operaciones')
plt.xlabel('nanosegundos')
plt.ylabel('Fidelidad')
```

```python
#Mean time
dfGraf = pd.concat([df21['Mean time'],df22['Mean time'],
                                         df23['Mean time']],axis=1)
dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Tiempo medio entrelazamiento frente \n tiempo de ejecución operaciones')
plt.xlabel('nanosegundos')
plt.ylabel('nanosegundos')
```

```python
#Mean time
dfGraf = pd.concat([df21['EPS'],df22['EPS'],
                                         df23['EPS']],axis=1)
dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Capacidad respecto tiempo operaciones')
plt.xlabel('nanosegundos')
plt.ylabel('entrelazamientos por segundo')
```
