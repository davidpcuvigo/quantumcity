```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df38a = pd.read_csv('prueba38a.csv',delimiter=',',index_col=3)
df38b = pd.read_csv('prueba38b.csv',delimiter=',',index_col=3)
df38c = pd.read_csv('prueba38c.csv',delimiter=',',index_col=3)
df38d = pd.read_csv('prueba38d.csv',delimiter=',',index_col=3)
df38e = pd.read_csv('prueba38e.csv',delimiter=',',index_col=3)
```

```python
#Fidelity
dfGraf = pd.concat([df38a['Mean fidel'],df38b['Mean fidel'],
                                         df38c['Mean fidel'],df38d['Mean fidel'],
                                        df38e['Mean fidel']],axis=1)
dfGraf.columns=['10 ns','1000 ns','5000 ns','15000 ns','30000 ns']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['10 ns'].plot(color='b',legend='True')
_ = dfGraf['1000 ns'].plot(color='r',legend='True')
_ = dfGraf['5000 ns'].plot(color='g',legend='True')
_ = dfGraf['15000 ns'].plot(color='y',legend='True')
_ = dfGraf['30000 ns'].plot(color='m',legend='True')
plt.title('Evolucin de la fidelidad con la tasa de cambio de \n fase en procesadores y duracin de instrucciones')
plt.xlabel('Tasa de dephase (Hz)')
plt.ylabel('Fidelidad')
```

```python
#Mean time
dfGraf = pd.concat([df38a['Mean time'],df38b['Mean time'],
                                         df38c['Mean time'],df38d['Mean time'],
                                        df38e['Mean time']],axis=1)
dfGraf.columns=['10 ns','1000 ns','5000 ns','15000 ns','30000 ns']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['10 ns'].plot(color='b',legend='True')
_ = dfGraf['1000 ns'].plot(color='r',legend='True')
_ = dfGraf['5000 ns'].plot(color='g',legend='True')
_ = dfGraf['15000 ns'].plot(color='y',legend='True')
_ = dfGraf['30000 ns'].plot(color='m',legend='True')
plt.title('Tiempo de generacin de entrelazamiento con la tasa de cambio de \n fase en procesadores y duracin de instrucciones')
plt.xlabel('Tasa de dephase (Hz)')
plt.ylabel('nanosegundos')
```
