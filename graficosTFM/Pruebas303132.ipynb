```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df30 = pd.read_csv('prueba30.csv',delimiter=',',index_col=3)
df31 = pd.read_csv('prueba31.csv',delimiter=',',index_col=3)
df32 = pd.read_csv('prueba32.csv',delimiter=',',index_col=3)
dfmax = pd.read_csv('max-teleport.pruebas30-35.csv',delimiter=',',index_col=3)
```

```python
#Fidelity
dfGraf = pd.concat([df30['Mean fidel'],df31['Mean fidel'],
                                         df32['Mean fidel']],axis=1)

dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolución de la fidelidad con el tiempo \n de ejecución instrucciones')
plt.xlabel('nanosegundos')
plt.ylabel('Fidelidad')
```

```python
#Mean time
dfGraf = pd.concat([df30['Mean time'],df31['Mean time'],
                                         df32['Mean time']],axis=1)

dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolución de tiempo medio con el tiempo de instrucciones')
plt.xlabel('nanosegundos')
plt.ylabel('nanosegundos')
```

```python
#Teleported states
dfGraf = pd.concat([df30['Teleported states'],df31['Teleported states'],
                                         df32['Teleported states'],dfmax['Teleported states']],axis=1)

dfGraf.columns=['1 switch','3 switches','5 switches','limite']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
_ = dfGraf['limite'].plot(color='k',linestyle='dashed')
plt.axis([10, 100000, 85, 150])
plt.title('Evolución de estados teleportados con el \n tiempo de ejecución de instrucciones')
plt.xlabel('nanosegundos')
plt.ylabel('estados teleportados')
```

```python
#Qubits descartados: ninguno
dfGraf = pd.concat([df30['Discarded qubits'],df31['Discarded qubits'],
                                         df32['Discarded qubits']],axis=1)

dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolución dqubits descartados con el tiempo de instrucciones')
plt.xlabel('estados teleportados')
plt.ylabel('nanosegundos')
```

```python
#Tamaño cola: ninguno
dfGraf = pd.concat([df30['Queue size'],df31['Queue size'],
                                         df32['Queue size']],axis=1)

dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolución tamaño cola con el tiempo de \n ejecución de instrucciones')
plt.xlabel('nanosegundos')
plt.ylabel('qubits encolados')
```
