```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df33 = pd.read_csv('prueba33.csv',delimiter=',',index_col=3)
df34 = pd.read_csv('prueba34.csv',delimiter=',',index_col=3)
df35 = pd.read_csv('prueba35.csv',delimiter=',',index_col=3)
dfmax = pd.read_csv('max-teleport.pruebas30-35.csv',delimiter=',',index_col=3)
```

```python
#Fidelity
dfGraf = pd.concat([df33['Mean fidel'],df34['Mean fidel'],
                                         df35['Mean fidel']],axis=1)

dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolución de la fidelidad con el tiempo de \n ejecución de instrucciones')
plt.xlabel('nanosegundos')
plt.ylabel('Fidelidad')
```

```python
#Mean time
dfGraf = pd.concat([df33['Mean time'],df34['Mean time'],
                                         df35['Mean time']],axis=1)

dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolución de tiempo medio con el tiempo de instrucciones')
plt.xlabel('nanosegundos')
plt.ylabel('nanosegundos')
```

```python
#Teleported states
dfGraf = pd.concat([df33['Teleported states'],df34['Teleported states'],
                                         df35['Teleported states'],dfmax['Teleported states']],axis=1)

dfGraf.columns=['1 switch','3 switches','5 switches','limite']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
_ = dfGraf['limite'].plot(color='k',linestyle='dashed')
plt.axis([10, 100000, 85, 350])
plt.title('Evolución de estados teleportados con el tiempo de \n ejecución de instrucciones')
plt.xlabel('nanosegundos')
plt.ylabel('estados teleportados')
```

```python
#Qubits descartados: ninguno
dfGraf = pd.concat([df33['Discarded qubits'],df34['Discarded qubits'],
                                         df35['Discarded qubits']],axis=1)

dfGraf.columns=['1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolución dqubits descartados con el tiempo de instrucciones')
plt.xlabel('estados teleportados')
plt.ylabel('nanosegundos')
```
