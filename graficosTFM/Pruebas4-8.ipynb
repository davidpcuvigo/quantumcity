```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df1 = pd.read_csv('prueba4.csv',delimiter=',',index_col=3)
df2 = pd.read_csv('prueba8.csv',delimiter=',',index_col=3)
```

# Tiempo medio de espera por solicitud en función del material de la memoria
Aunque en  general el NV va mejor que el SiV, como las puertas de 1 qubit son más rápidas en el SiV, cuando no hay purificación (no se usan puertas de 2 qubits), el SiV mejora bastante el tiempo de las solicitudes

```python

# Plot the results
plt.figure(figsize=(6, 3))
plt.plot(df1['Mean time'], color='r',label='NV')
plt.plot(df2['Mean time'], color='g', label='SiV') #mira lo que pasa si multiplicas los valores por 4 (no he encontrado el porque)
plt.xlabel('Probabilidad inicial de pérdida de fotón')
plt.ylabel('Duración media de la solicitud (ns)')
plt.title('Evolución del tiempo de generación de entrelazamiento \n con $p_{loss init}$')
plt.legend()
#plt.grid(True)
plt.show()
```
