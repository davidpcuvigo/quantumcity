```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df1 = pd.read_csv('prueba11.csv',delimiter=',',index_col=3)
df2 = pd.read_csv('prueba12.csv',delimiter=',',index_col=3)
df3 = pd.read_csv('prueba13.csv',delimiter=',',index_col=3)
df4 = pd.read_csv('prueba14.csv',delimiter=',',index_col=3)

df1.index=df1.index*2 +1
df2.index=df2.index*4 +1
df3.index=df3.index*2 +1
df4.index=df4.index*4 +1
```

# EPRs generadas respecto la distancia para cada tecnología
Aquí es gracioso ver la immensa diferencia entre NV y SiV, que se debe solamente a la duración de la puerta CX. Realmente me hace plantearme si la referencia aquella estaba bien.

```python
# Plot the results
plt.figure(figsize=(6, 3))
plt.plot(df1['Generated entanglements'], color='r',label='3 switches NV')
plt.plot(df2['Generated entanglements'], color='g', label='5 switches NV')
plt.plot(df3['Generated entanglements'], color='b',label='3 switches SiV')
plt.plot(df4['Generated entanglements'], color='y', label='5 switches SiV')


plt.xlabel('Distancias de los nodos finales (km)')
plt.ylabel('EPRs generadas')
plt.title('Capacidad en función de la distancia entre end nodes (varian switches)')
plt.legend()
#plt.grid(True)
plt.show()
```

# Tradeoff entre capacidad y fidelidad
En este gráfico con el inset se ve como en el rango distancias donde la capacidad de 3 switches es mucho mayor, la fidelidad es estadísticamente la misma, en cambio, en distancias largas (>20 km), las capacidades ya son similares y en cambio la fidelidad para 3 switches es (un poco) peor. Así que se puede concluir que para distancias cortas son mejor 3 switches y para largas 5.

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df1['Generated entanglements'], color='r', label='3 switches (Generated entanglements)')
ax.plot(df2['Generated entanglements'], color='g', label='5 switches (Generated entanglements)')

# Add labels, title, and legend
ax.set_xlabel('Distancias de los nodos finales (km)')
ax.set_ylabel('EPRs generadas')
ax.set_title('Capacidad en función de la distancia entre end nodes (NV)')
ax.legend()

# Create the inset plot
inset_ax = inset_axes(ax, width="40%", height="50%", loc="right")  # adjust the size and location as needed
inset_ax.plot(df1['Mean fidel'], color='r', linestyle='', marker='.', label='3 switches')
inset_ax.plot(df2['Mean fidel'], color='g', linestyle='', marker='.', label='5 switches')
inset_ax.set_xlabel('Distancias de los nodos finales (km)')
inset_ax.set_ylabel('Fidelidad')
inset_ax.set_title('Fidelidad media')
inset_ax.legend()
```
