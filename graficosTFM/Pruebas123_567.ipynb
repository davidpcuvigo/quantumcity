```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df1 = pd.read_csv('prueba1.csv',delimiter=',',index_col=3)
df2 = pd.read_csv('prueba2.csv',delimiter=',',index_col=3)
df3 = pd.read_csv('prueba3.csv',delimiter=',',index_col=3)
df5 = pd.read_csv('prueba5.csv',delimiter=',',index_col=3)
df6 = pd.read_csv('prueba6.csv',delimiter=',',index_col=3)
df7 = pd.read_csv('prueba7.csv',delimiter=',',index_col=3)
```

# Fidelidad media respecto al número de solicitudes para cada tecnología

```python
average_fidelity1 = df1.groupby('valor')['Mean fidel'].mean().reset_index() #innecesario
average_fidelity2 = df2.groupby('valor')['Mean fidel'].mean().reset_index()
average_fidelity3 = df3.groupby('valor')['Mean fidel'].mean().reset_index()

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(average_fidelity1['Mean fidel'], color='r',label='1 Request')
plt.plot(average_fidelity2['Mean fidel'], color='g', label='3 Requests')
plt.plot(average_fidelity3['Mean fidel'], color='b', label='5 Requests')
plt.xlabel('Distancias de los nodos finales (km)')
plt.ylabel('Fidelidad')
plt.title('Fidelidad media de las solicitudes (NV)')
plt.legend()
#plt.grid(True)
plt.show()
```

```python
average_fidelity5 = df5.groupby('valor')['Mean fidel'].mean().reset_index() #innecesario
average_fidelity6 = df6.groupby('valor')['Mean fidel'].mean().reset_index()
average_fidelity7 = df7.groupby('valor')['Mean fidel'].mean().reset_index()

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(average_fidelity5['Mean fidel'], color='r',label='1 Request')
plt.plot(average_fidelity6['Mean fidel'], color='g', label='3 Requests')
plt.plot(average_fidelity7['Mean fidel'], color='b', label='5 Requests')
plt.xlabel('Distancias de los nodos finales (km)')
plt.ylabel('Fidelidad')
plt.title('Fidelidad media de las solicitudes (SiV)')
plt.legend()
#plt.grid(True)
plt.show()
```

# Comparación de tecnologías
## No se si queda mejor con el rango de 0 a 1 o los originales, porque la verdad es que el SiV se ve bastante atroz. Ya lo hablaremos

```python
# Plot Fidelidad Media vs Distancia
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
plt.plot(average_fidelity1['Mean fidel'], color='r', label='NV')
plt.plot(average_fidelity5['Mean fidel'], color='g', label='SiV')
plt.xlabel('Distancias de los nodos finales (km)')
plt.ylabel('Fidelidad')
plt.title('Fidelidad media de la solicitud')
plt.legend()
plt.ylim(0,1)

plt.subplot(1, 3, 2)
plt.plot(average_fidelity2['Mean fidel'], color='r', label='NV')
plt.plot(average_fidelity6['Mean fidel'], color='g', label='SiV')
plt.xlabel('Distancias de los nodos finales (km)')
plt.ylabel('Fidelidad')
plt.title('Fidelidad media de las 3 solicitudes')
plt.legend()
plt.ylim(0,1)

plt.subplot(1, 3, 3)
plt.plot(average_fidelity3['Mean fidel'], color='r', label='NV')
plt.plot(average_fidelity7['Mean fidel'], color='g', label='SiV')
plt.xlabel('Distancias de los nodos finales (km)')
plt.ylabel('Fidelidad')
plt.title('Fidelidad media de las 5 solicitudes (SiV)')
plt.legend()
plt.ylim(0,1)

plt.tight_layout()
plt.show()
```
