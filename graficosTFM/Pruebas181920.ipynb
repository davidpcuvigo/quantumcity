```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df18 = pd.read_csv('prueba18.csv',delimiter=',')
df19 = pd.read_csv('prueba19.csv',delimiter=',')
df20 = pd.read_csv('prueba20.csv',delimiter=',')

```

```python
#Fidelity
dfGraf = pd.concat([df18['Mean fidel'],df19['Mean fidel'],
                                         df20['Mean fidel']],axis=1)
dfGraf.reset_index(inplace=True)
dfGraf['Valor'] = dfGraf['index']*3.878 + 10
dfGraf.set_index('Valor',inplace=True)
dfGraf.columns=['rowold','1 switch','3 switches','5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 switch'].plot(color='b',legend='True')
_ = dfGraf['3 switches'].plot(color='r',legend='True')
_ = dfGraf['5 switches'].plot(color='g',legend='True')
plt.title('Evolucin de la fidelidad con la distancia extremo-extremo')
plt.xlabel('km')
plt.ylabel('Fidelidad')
```
