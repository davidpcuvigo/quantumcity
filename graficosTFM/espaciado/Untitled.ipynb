```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df15a = pd.read_csv('prueba15a.csv',delimiter=',',index_col=3)
df15b = pd.read_csv('prueba15b.csv',delimiter=',',index_col=3)
df16a = pd.read_csv('prueba16a.csv',delimiter=',',index_col=3)
df16b = pd.read_csv('prueba16b.csv',delimiter=',',index_col=3)
df17a = pd.read_csv('prueba17a.csv',delimiter=',',index_col=3)
df17b = pd.read_csv('prueba17b.csv',delimiter=',',index_col=3)
```

```python
df17a.info()
```

```python
#Fidelity
dfGraf = pd.concat([df15a['Mean fidel'],df15b['Mean fidel'],
                                         df16a['Mean fidel'],df16b['Mean fidel'],
                                        df17a['Mean fidel'],df17b['Mean fidel']],axis=1)
dfGraf.columns=['10km - 1 switch','100km - 1 switch','10km - 3 switches','100km - 3 switches','10km - 5 switches','100km - 5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['10km - 1 switch'].plot(color='b',linestyle='dashed',legend='True',logx=True)
_ = dfGraf['100km - 1 switch'].plot(color='b',legend='True',logx=True)
_ = dfGraf['10km - 3 switches'].plot(color='r',linestyle='dashed',legend='True',logx=True)
_ = dfGraf['100km - 3 switches'].plot(color='r',legend='True',logx=True)
_ = dfGraf['10km - 5 switches'].plot(color='g',linestyle='dashed',legend='True',logx=True)
_ = dfGraf['100km - 5 switches'].plot(color='g',legend='True',logx=True)
plt.title('Fidelity evolution with T1 memory time')
plt.xlabel('T1 (nanoseconds)')
plt.ylabel('Fidelity')
```

```python
#Fidelity
dfGraf = pd.concat([df15a['Mean time'],df15b['Mean time'],
                                         df16a['Mean time'],df16b['Mean time'],
                                        df17a['Mean time'],df17b['Mean time']],axis=1)
dfGraf.columns=['10km - 1 switch','100km - 1 switch','10km - 3 switches','100km - 3 switches','10km - 5 switches','100km - 5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['10km - 1 switch'].plot(color='b',linestyle='dashed',legend='True')
_ = dfGraf['100km - 1 switch'].plot(color='b',legend='True')
_ = dfGraf['10km - 3 switches'].plot(color='r',linestyle='dashed',legend='True')
_ = dfGraf['100km - 3 switches'].plot(color='r',legend='True')
_ = dfGraf['10km - 5 switches'].plot(color='g',linestyle='dashed',legend='True')
_ = dfGraf['100km - 5 switches'].plot(color='g',legend='True')
plt.title('Entanglement generation time evolution with T1 memory time')
plt.xlabel('T1 (nanoseconds)')
plt.ylabel('Mean time (nanoseconds)')
```

```python
#Fidelity
dfGraf = pd.concat([df15a['EPS'],df15b['EPS'],
                                         df16a['EPS'],df16b['EPS'],
                                        df17a['EPS'],df17b['EPS']],axis=1)
dfGraf.columns=['10km - 1 switch','100km - 1 switch','10km - 3 switches','100km - 3 switches','10km - 5 switches','100km - 5 switches']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['10km - 1 switch'].plot(color='b',linestyle='dashed',legend='True')
_ = dfGraf['100km - 1 switch'].plot(color='b',legend='True')
_ = dfGraf['10km - 3 switches'].plot(color='r',linestyle='dashed',legend='True')
_ = dfGraf['100km - 3 switches'].plot(color='r',legend='True')
_ = dfGraf['10km - 5 switches'].plot(color='g',linestyle='dashed',legend='True')
_ = dfGraf['100km - 5 switches'].plot(color='g',legend='True')
plt.title('Entanglement generation rate evolution with T1 memory time')
plt.xlabel('T1 (nanoseconds)')
plt.ylabel('EPRs per second')
```
