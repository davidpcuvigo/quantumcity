```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df41a = pd.read_csv('prueba41a.csv',delimiter=',',index_col=3)
df41b = pd.read_csv('prueba41b.csv',delimiter=',',index_col=3)
df41c = pd.read_csv('prueba41c.csv',delimiter=',',index_col=3) #Este tiene fidelidad sin supresión alguna DepolarGauss
df41d = pd.read_csv('prueba41d.csv',delimiter=',',index_col=3) #Este tiene fidelidad sin supresión alguna Dephase
df42a = pd.read_csv('prueba42a.csv',delimiter=',',index_col=3)
df42b = pd.read_csv('prueba42b.csv',delimiter=',',index_col=3)
```

# Con error en canal de Dephase (Z)

```python
#Fidelity
dfGraf = pd.concat([df41a['Mean fidelity'],df42a['Mean fidelity'],
                                         df41d['Mean fidelity']],axis=1)
dfGraf.columns=['Purificación','QEC','Sin corrección']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['Purificación'].plot(color='b',legend='True')
_ = dfGraf['QEC'].plot(color='r',legend='True')
_ = dfGraf['Sin corrección'].plot(color='k',linestyle='dashed',legend='True')

plt.title('Evolución de la fidelidad con distancia \n aplicando técnicas de supresión')
plt.xlabel('Distancia enlace (km)')
plt.ylabel('Fidelidad')
```

```python
#Teleported states
dfGraf = pd.concat([df41a['Teleported states'],df42a['Teleported states'],
                                         df41d['Teleported states']],axis=1)
dfGraf.columns=['Purificación','QEC(Shor)','Sin corrección']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['Purificación'].plot(color='b',legend='True')
_ = dfGraf['QEC(Shor)'].plot(color='r',legend='True')

ax = plt.gca()
ax.set_xlim([0, 100])
ax.set_ylim([0, 300])
plt.title('Número de estados teleportados en 1 segundo')
plt.xlabel('Distancia enlace (km)')
plt.ylabel('#qubits')
```

```python
#Mean time
dfGraf = pd.concat([df41a['Mean time'],df42a['Mean time'],
                                         df41d['Mean time']],axis=1)
dfGraf.columns=['Purificación','QEC(Shor)','Sin corrección']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['Purificación'].plot(color='b',legend='True')
_ = dfGraf['QEC(Shor)'].plot(color='r',legend='True')

plt.title('Tiempo medio de teleportación de un qubit')
plt.xlabel('Distancia enlace (km)')
plt.ylabel('nanosegundos')
```
