```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

df24 = pd.read_csv('prueba24.csv',delimiter=',')
df24.set_index('distance',inplace=True)

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(projection = '3d')
ax.set_xlabel('km',labelpad=10)
ax.set_ylabel('nanoseconds',labelpad=20)
ax.set_zlabel('eps',labelpad=20)
ax.set_title('Capacidad frente a distancia extremo a extremo \n y duracin e instrucciones')

for time in [10,1000,5000,15000,30000]:
    ax.plot_trisurf(df24[df24['gate time']==time].index,df24[df24['gate time']==time]['valor'],df24[df24['gate time']==time]['EPS'],linewidth=0.2, antialiased=True)
    ax.view_init(elev=10., azim=275)
plt.legend(['10ns','1000ns','5000ns','15000ns','30000ns'],loc="upper right")

plt.show()

#Lo que se ve: Para valores bajos e distancias, 
#el tiempo de ejecucin en puertas incluye de forma decisiva en capacidad.
#No as la decoherencia, la capacidad es constante
```
