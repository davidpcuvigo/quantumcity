```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df25 = pd.read_csv('prueba25.csv',delimiter=',',index_col=3)
df26 = pd.read_csv('prueba26.csv',delimiter=',',index_col=3)
df27 = pd.read_csv('prueba27.csv',delimiter=',',index_col=3)
df28 = pd.read_csv('prueba28.csv',delimiter=',',index_col=3)
df29 = pd.read_csv('prueba29.csv',delimiter=',',index_col=3)
R1EPS26 = df26[df26['request']=='request1']['EPS']
R2EPS26 = df26[df26['request']=='request2']['EPS']
R1EPS27 = df27[df27['request']=='request1']['EPS']
R2EPS27 = df27[df27['request']=='request2']['EPS']
R3EPS27 = df27[df27['request']=='request3']['EPS']
R1EPS28 = df28[df28['request']=='request1']['EPS']
R2EPS28 = df28[df28['request']=='request2']['EPS']
R3EPS28 = df28[df28['request']=='request3']['EPS']
R4EPS28 = df28[df28['request']=='request4']['EPS']
R1EPS29 = df29[df29['request']=='request1']['EPS']
R2EPS29 = df29[df29['request']=='request2']['EPS']
R3EPS29 = df29[df29['request']=='request3']['EPS']
R4EPS29 = df29[df29['request']=='request4']['EPS']
R5EPS29 = df29[df29['request']=='request5']['EPS']
EPS25 = df25['EPS']
EPS26 = R1EPS26+R2EPS26
EPS27 = R1EPS27+R2EPS27+R3EPS27
EPS28 = R1EPS28+R2EPS28+R3EPS28+R4EPS28
EPS29 = R1EPS29+R2EPS29+R3EPS29+R4EPS29+R5EPS29

```

```python
#Capacity
dfGraf = pd.concat([EPS25,EPS26,EPS27,EPS28,EPS29],axis=1)

dfGraf.columns=['1 request','2 requests','3 requests','4 requests','5 requests']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 request'].plot(color='b',legend='True')
_ = dfGraf['2 requests'].plot(color='r',legend='True')
_ = dfGraf['3 requests'].plot(color='g',legend='True')
_ = dfGraf['4 requests'].plot(color='y',legend='True')
_ = dfGraf['5 requests'].plot(color='m',legend='True')
plt.title('Capacity evolution with gate duration')
plt.xlabel('Gate duration (ns)')
plt.ylabel('Capacity (EPR/s)')
plt.xscale('log')
plt.savefig("cap_gates.pdf", format="pdf", bbox_inches="tight")
```

```python
#Capacity
dfGraf = pd.concat([EPS25,EPS26,EPS27,EPS28,EPS29],axis=1)

dfGraf.columns=['1 petición','2 peticiones','3 peticiones','4 peticiones','5 peticiones']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 petición'].plot(color='b',legend='True',logx=True)
_ = dfGraf['2 peticiones'].plot(color='r',legend='True',logx=True)
_ = dfGraf['3 peticiones'].plot(color='g',legend='True',logx=True)
_ = dfGraf['4 peticiones'].plot(color='y',legend='True',logx=True)
_ = dfGraf['5 peticiones'].plot(color='m',legend='True',logx=True)
plt.title('Capacidad total del switch frente a duración \n de instrucciones')
plt.xlabel('nanosegundos')
plt.ylabel('entrelazamientos por segundo')
```

```python
#Mean time
dfGraf = pd.concat([df25['Mean time'],df26[df26['request']=='request1']['Mean time'],
                    df27[df27['request']=='request1']['Mean time'],
                   df28[df28['request']=='request1']['Mean time'],
                   df29[df29['request']=='request1']['Mean time']],
                   axis=1)
dfGraf.columns=['1 petición','2 peticiones','3 peticiones','4 peticiones','5 peticiones']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 petición'].plot(color='b',legend='True',logx=True)
_ = dfGraf['2 peticiones'].plot(color='r',legend='True',logx=True)
_ = dfGraf['3 peticiones'].plot(color='g',legend='True',logx=True)
_ = dfGraf['4 peticiones'].plot(color='y',legend='True',logx=True)
_ = dfGraf['5 peticiones'].plot(color='m',legend='True',logx=True)
plt.title('Tiempo medio de generación de entrelazamiento \n frente a duración de instrucciones')
plt.xlabel('nanosegundos')
plt.ylabel('nanosegundos')
```

```python
#Mean fidelity
dfGraf = pd.concat([df25['Mean fidel'],df26[df26['request']=='request1']['Mean fidel'],
                    df27[df27['request']=='request1']['Mean fidel'],
                   df28[df28['request']=='request1']['Mean fidel'],
                   df29[df29['request']=='request1']['Mean fidel']],
                   axis=1)
dfGraf.columns=['1 petición','2 peticiones','3 peticiones','4 peticiones','5 peticiones']
fig = plt.figure(figsize=(6,3))
_ = dfGraf['1 petición'].plot(color='b',legend='True')
_ = dfGraf['2 peticiones'].plot(color='r',legend='True')
_ = dfGraf['3 peticiones'].plot(color='g',legend='True')
_ = dfGraf['4 peticiones'].plot(color='y',legend='True')
_ = dfGraf['5 peticiones'].plot(color='m',legend='True')
plt.title('Fielidad en función de tiempo de instrucciones')
plt.xlabel('Fidelidad')
plt.ylabel('nanosegundos')
```
