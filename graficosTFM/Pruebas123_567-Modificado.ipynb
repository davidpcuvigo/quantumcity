```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df1 = pd.read_csv('prueba1.csv',delimiter=',',index_col=3)
df2 = pd.read_csv('prueba2.csv',delimiter=',',index_col=3)
df3 = pd.read_csv('prueba3.csv',delimiter=',',index_col=3)
df5 = pd.read_csv('prueba5.csv',delimiter=',',index_col=3)
df6 = pd.read_csv('prueba6.csv',delimiter=',',index_col=3)
df7 = pd.read_csv('prueba7.csv',delimiter=',',index_col=3)
```

# Fidelidad media respecto al número de solicitudes para cada tecnología

```python
average_fidelity1 = df1.groupby('valor')['Mean fidel'].mean().reset_index() #innecesario
average_fidelity2 = df2.groupby('valor')['Mean fidel'].mean().reset_index()
average_fidelity3 = df3.groupby('valor')['Mean fidel'].mean().reset_index()
average_fidelity1.set_index('valor',inplace=True)
average_fidelity2.set_index('valor',inplace=True)
average_fidelity3.set_index('valor',inplace=True)
# Plot the results
plt.figure(figsize=(6, 3))
plt.plot(average_fidelity1['Mean fidel'], color='r',label='1 Request')
plt.plot(average_fidelity2['Mean fidel'], color='g', label='3 Requests')
plt.plot(average_fidelity3['Mean fidel'], color='b', label='5 Requests')
plt.xlabel('Distancias de enlaces (km)')
plt.ylabel('Fidelidad')
plt.title('Fidelidad media de las solicitudes (NV)')
plt.legend()
#plt.grid(True)
plt.show()
```

```python
#Tiempos
average_fidelity1 = df1.groupby('valor')['Mean time'].mean().reset_index() #innecesario
average_fidelity2 = df2.groupby('valor')['Mean time'].mean().reset_index()
average_fidelity3 = df3.groupby('valor')['Mean time'].mean().reset_index()
average_fidelity1.set_index('valor',inplace=True)
average_fidelity2.set_index('valor',inplace=True)
average_fidelity3.set_index('valor',inplace=True)
# Plot the results
plt.figure(figsize=(6, 3))
plt.plot(average_fidelity1['Mean time'], color='r',label='1 Request')
plt.plot(average_fidelity2['Mean time'], color='g', label='3 Requests')
plt.plot(average_fidelity3['Mean time'], color='b', label='5 Requests')
plt.xlabel('Distancia de enlaces (km)')
plt.ylabel('Nanosegundos')
plt.title('Tiempo medio de generación de entrelazamiento \n de las solicitudes (NV)')
plt.legend()
#plt.grid(True)
plt.show()
```
